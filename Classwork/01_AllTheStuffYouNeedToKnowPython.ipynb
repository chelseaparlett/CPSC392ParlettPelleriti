```python
# import necessary packages here
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
```

```python
# load World Happiness Report
world_happiness = pd.read_csv("/content/drive/MyDrive/2015.csv")
```

```python
# grap first 10 rows
world_happiness.head(10)
```

```python
# column info
world_happiness.info()
```

```python
# summary of continuous variables
world_happiness.describe()
```

```python
# print out column names
world_happiness.columns
```

```python
# get 10 countries with higest happiness rank
world_happiness.groupby("Country").mean().nlargest(10, "Happiness Rank")
```

```python
# loop through columns
for col in world_happiness:
  print(col)
```

```python
# loop through rows
for i in range(world_happiness.shape[0]):
  print(world_happiness.iloc[i])
```

```python
# use regular expressions to filter columns
# world_happiness.filter(regex = '.*\(.+\)', axis = 1)
world_happiness.filter(regex = 'Happiness', axis = 1)
```

```python
# grab boolean column
bool_we = world_happiness["Region"] == "Western Europe"

# use bools to filter dataframe
we = world_happiness.loc[bool_we]

# print head
we.head()
```
