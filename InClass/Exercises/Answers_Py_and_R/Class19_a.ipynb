1. Using the same breast cancer dataset we looked at in class, try different k's (2,3,4 and 5). Which has the best silhouette score? What does that indicate?

2. How did this perform compared to the k-means algorithm?

```python
bc = pd.read_csv("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/BreastCancer.csv")
print(bc.shape)
print(bc.columns)

bc.head()
```

```python
feat = bc.columns[3:10]

X = bc[feat]
z = StandardScaler()
X[feat] = z.fit_transform(X)

ks = [2,3,4,5]
members = {}
for k in ks:
    km = GaussianMixture(k)
    km.fit(X)
    memebership = km.predict(X)
    
    members[k] = memebership
    
    print("k = ", k, "silhouette: ", silhouette_score(X, memebership))
    
bc["best"] = members[2]
```
